```python
data = {
    "salary": [10000, 20200, 35000, 45000, 6000, 67000, 78000, 89000, 91000, 10200]
}
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
#line plot
plt.plot(df["salary"], color = "black" ,marker='o',linestyle=":",linewidth="2")
plt.grid()
plt.show()
```

```python
#Histogram
plt.hist(df["salary"],bins=5,color="pink")
plt.show()
```

```python
#boxplot
plt.boxplot(df["salary"])
plt.show()
```

```python
df["dept"]=[ "HR",
        "Finance",
        "IT",
        "Marketing",
        "Sales",
        "IT",
        "Engineering",
        "HR",
        "Finance",
        "HR"]
df.head()
```

```python
#pie chart
count=df["dept"].value_counts()
count
```

```python
explode = [0.1 if i == 0 else 0 for i in range(len(count))]
plt.pie(count,labels=count.index, autopct="%1.1f", explode=explode)
plt.axis("equal")
plt.show()
```

```python
#countplot
plt.bar(count.index,count,color=["lightpink","skyblue","lightgreen"])
plt.show()
```

```python
#comparision two data by plot
df["Age"]=[25, 30, 28, 35, 32, 27, 40, 29, 31, 38]
df.head()
```

```python
#Scatter plot
plt.scatter(df["salary"], df["Age"], color="orange")
plt.show()
```
